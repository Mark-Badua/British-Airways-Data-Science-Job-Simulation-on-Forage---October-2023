```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
df = pd.read_csv("BA_reviews_verified.csv", index_col = 0)
df['reviews'] = df.reviews.apply(str)
```

```python
comment_words = ""
stopwords = set(STOPWORDS)
 
# iterate through the csv file
for val in df.reviews:
    # split the value -- create list of words
    tokens = val.split()
     
    # Converts each token into lowercase
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
     
    comment_words += " ".join(tokens)+" "
```

```python
wordcloud = WordCloud(stopwords = stopwords, min_font_size =10 ,background_color = 'white').generate(comment_words)
 
```

```python
plt.figure(figsize = (10,6), dpi = 600)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.savefig('BA_flightreview_wordcloud.png', dpi = 600, )
```

**Insight**

Most of the reviews are associated with seat, food, time, service, cabin crew, luggage, flight, business class, passenger, plane, meal, experience, drink, bag, cabin, lounge, london, heathrow
